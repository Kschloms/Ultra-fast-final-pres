```python
from src.Double_integral_method  import M
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import scipy.integrate as integrate
```

```python
#Field parameters
Nc = 2 #Field revolutions in envelope (Cycles during pulse duration)
eps = 0 #(-1,0,1), Polarization of light
phi = 0 #Phase difference between envelope and field
#all values in au
wl = 0.057 #800nm wavelength
I = 0.003 #I=10^14 w/cm^2  
E0 = 0.500 #13.6 eV
 

M1 = M(Nc,eps,phi,wl,I,E0)
```

```python
k_zs = np.linspace(-1,1,100) 
k_xs = np.zeros_like(k_zs)
k_ys = np.zeros_like(k_zs)

# Create an array of shape (100, 3) where each row is a k vector [kx, ky, kz]
ks = np.column_stack((k_xs, k_ys, k_zs))
Mk_squared_vals = M1.Mk0_squared(ks)
```

```python
fig, ax = plt.subplots()
# Plot Mk_squared against k_z
ax.plot(k_zs, Mk_squared_vals)
ax.set_xlabel('k_z')
ax.set_ylabel('Mk_squared')
ax.set_title('Mk_squared vs k_z')
ax.grid(True)
```
